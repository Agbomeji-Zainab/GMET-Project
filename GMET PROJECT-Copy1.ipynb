Building a Decision Tree Regressor Model to Predict Calories Burned

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Encode categorical variables
le_gender = LabelEncoder()
le_workout_type = LabelEncoder()


data['Gender'] = le_gender.fit_transform(data['Gender'])
data['Workout_Type'] = le_workout_type.fit_transform(data['Workout_Type'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
```

```python
#Splitting Data into X and Y
X = data.drop(columns='Calories_Burned')
y = data['Calories_Burned']
X.shape, y.shape
```

```python
#Spliting the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
#Initialize Decision Tree Regressor
dtr = DecisionTreeRegressor(max_depth = 2)
```

```python
#Train the model
dtr.fit(X_train, y_train)
```

```python
from sklearn.tree import export_text
```

```python
print(export_text(dtr, feature_names=X.columns.to_list()))
```

```python
#Make predictions
y_pred = dtr.predict(X_test)
```

```python
#Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
#Calculate RMSE
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")
```

```python
#Calculate the range of Calories Burned (actual values)
min_calories = y_test.min()  # Minimum value in test set
max_calories = y_test.max()  # Maximum value in test set
print(f"Range of Calories Burned: {min_calories} to {max_calories}")
```

```python
from sklearn.metrics import r2_score
```

```python
#Calculate R-squared
r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
```

In this project, I built a Decision Tree Regressor to predict Calories Burned based on several features like Age, Weight, Height, Workout Frequency, and more. The model was trained and tested on a dataset to explore how these features influence calories burned during physical activity.

An R-squared value of 0.76 means my model accounts for 76.3% of the variation in the target variable, 'Calories Burned'. This signifies that the model has captured a strong relationship between the input features and the target. Usually, a real-world data performance with an R-squared value of 0.76 can be considered solid, though higher values denote a better fit.
